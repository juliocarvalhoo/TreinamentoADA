```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Exemplos de amostragem.

Amostragem aleatória simples.

```python
#Com reposição
array = np.random.randint(1,50, size = (1,10))
display(array)
```

```python
#Sem reposição.

#criar array de 1 a 100
array = np.arange(1,101,1)
print("Array:",array)

#selecionar 10 itemns se repetição usamos a função sample.
amostra_sem_reposicao = np.random.choice(array,10, replace = True)
print("Amostra:",amostra_sem_reposicao)
```

Amostra estratificada

```python
#amostra é divida em estratos homogeneos, e selecionar um  grupo aleatório de cada estrato. Ex:

#criar array
array = np.arange(1,101,1) 

#estratificar entre os valores pares e ímpares
#relembrando uso de filter e lambda para selecionar os pares
arrayPar =  np.array(list(filter(lambda x : x % 2 ==0, array)))
#maneira mais facil, selecionar os valores ímpares
arrayImpar = array[array % 2 != 0]

#selecionar 10 elementos de cada estrato.
print("Amostra 1:",np.random.choice(arrayPar,10, replace= True))
print("Amostra 2:",np.random.choice(arrayImpar,10, replace = True))
```

Amostra Sistemática

```python
#Elaboro uma regra, ex: Selecionar 15 casas de um conjunto de 205 casas

array = np.arange(1,207,1)
#qual a regra vou seguir para selecionar os valores
#apos selecionar uma casa, selecionar outra a cada 14. (205/15) =~ 14
print("casas selecionadas:",array[::14])
```

Frequência

```python
listfrequencia = list(np.random.randint(low = 1, high =10, size = 30))
print(listfrequencia)
```

```python
#mostrando frequência
sns.set_style("darkgrid")
plt.title("Frequência dos numeros de 1 a 9")
frequencia = [listfrequencia.count(x) for x in listfrequencia]
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.bar(x = listfrequencia, height = frequencia)
plt.show()
```

frequência Relativa

```python
#Também há frequencia relativa acumulada.
frequencia_relativa = [listfrequencia.count(x) / len(listfrequencia) for x in listfrequencia]

plt.title("Frequência Relativa")
frequencia = [listfrequencia.count(x) for x in listfrequencia]
plt.xlabel('Valores')
plt.ylabel('Frequência Relativa')
plt.bar(x = listfrequencia, height = frequencia_relativa)
plt.show()
```

Histograma

```python
#Com o agrupamento dos valores proximos podemos criar um histograma

sns.histplot(listfrequencia)
plt.title("ex: Histograma")
plt.xticks(np.arange(0,11,2))
plt.xlabel('Valores intervalo')
plt.ylabel('Frequência')
plt.show()
```
