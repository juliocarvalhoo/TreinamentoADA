```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
distancia = [215,480,325,550,920,670,825,1070,1350,1250]
tempo_Entrega = [1,1,1.5,2,3,3,3.5,4,4.5,5]

dados_df = pd.DataFrame({"Distancia":distancia,"Tempo Entrega":tempo_Entrega})

dados_df
```

Gráfico de Dispersão

```python
sns.set_style("darkgrid")
sns.relplot(dados_df,x = "Distancia", y ="Tempo Entrega")
plt.title("Tempo x Distancia Entrega")
plt.show()
```

Analise de correlação

```python
correlação = dados_df["Distancia"].corr(dados_df["Tempo Entrega"])
correlação

#Exise uma correlação positiva muito grande!
```

Gráfico de regressão linear

```python
sns.scatterplot(dados_df, x = "Distancia", y = "Tempo Entrega")
sns.regplot(dados_df, x = "Distancia", y = "Tempo Entrega")
plt.show()
```

Calcular a reta de regressão

```python
# sabemos a reta é do formato y = mx + b
#Usaremos a biblioteca Numpy

valores = np.polyfit(dados_df["Distancia"],dados_df["Tempo Entrega"],1)
m = valores[0]
b = valores[1]

print(f"Y = {m:.5f}X + {b:.5f}")
```

Inserir a reta no primeiro gráfico

```python
x = np.array(dados_df["Distancia"])

sns.relplot(dados_df,x = "Distancia", y ="Tempo Entrega")
plt.plot(dados_df["Distancia"], m * x + b, color='red')

plt.title("Tempo x Distancia Entrega")
plt.show()
```

Coeficiente de determinação (R²)

correlação ao quadrado, mostra o quanto dos dados o gráfico pode explicar.

```python
CoeficienteDeterminacao = pow(correlação,2)
CoeficienteDeterminacao
```
