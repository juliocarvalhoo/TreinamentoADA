```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Limites

```python
#desafio, plotar gráfico de 1/x

plt.figure(figsize=(8,5))
plt.title("y = 1 / x")
x = np.linspace(-0.5,1,10000)
y = 1/x
plt.plot(x,y)
plt.show()

```

```python
plt.figure(figsize=(8,5))
plt.title("eixo positivo")
x = np.linspace(0.01,1,100)
y = 1/x
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("eixo negativo")
x = np.linspace(-1,-0.01,100)
y = 1/x
plt.plot(x,y)
plt.show()
```

##### Não existem limites Laterais

```python
'''Conceito de limite lateral'''

plt.figure(figsize=(8, 5))
plt.title("Limites laterais, tendendo para '+' e '-' infinito")

# Intervalo para x1 (valores negativos tendendo a zero)
x1 = np.linspace(-10, -0.01, 100)

# Intervalo para x2 (valores positivos tendendo a zero)
x2 = np.linspace(0.01, 10, 100)

y = 1 / x1
y2 = 1/ x2

plt.axhline(color="red", linestyle="dashed")  # linha horizontal em y=0
plt.axvline(color="red", linestyle="dashed")  # linha vertical em x=0
plt.plot(x1, y,label = "Eixo Negativo")
plt.plot(x2, y2, label = "Eixo Positivo")
plt.legend()

plt.show()
```
