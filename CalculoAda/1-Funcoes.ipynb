```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Para melhor compreensão do conteúdo das aulas, aqui é usado matplotlib e pyplot para representar os grafícos

```python
x = np.linspace(-10,10,100)
f = 2*x+1
fInversa = (x-1)/2
g = x**2
id = x
```

```python
"""FUNÇÃO INVERSA, EM RELAÇÃO A IDENTIDADE"""
plt.title("Função, Inversa e Identidade")
plt.plot(x,f, label = "f(x)")
plt.plot(x,id, label = "id")
plt.plot(x,fInversa, label = "inversa")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

plt.show()
```

```python
"""Composição de funções"""
gof = 4* pow(x,2) + (4*x) + 1
plt.title("Composição de funções")
plt.plot(x,f, label = "f(x)")
plt.plot(x,g, label = "g(x)")
plt.plot(x,gof, label = "Composta")


plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

plt.show()
```

```python
#função constante
fig = plt.figure( figsize=(8,5))
plt.title("Função Constante")

x = np.linspace(start = -10, stop = 10, num = 10000)
y = 20*np.ones(x.shape)

plt.plot(x,y)
plt.show()
```

```python
#função linear
plt.figure(figsize=(8,5))
plt.title("Função Linear")
x = np.linspace(-10,10,10000)
y = x + 2 
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Função Quadratica")
x=np.linspace(-20,20,10000)
y=pow(x,2)+2*x+1
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Função de Terceiro Grau")
x=np.linspace(-10,10,10000)
y=pow(x,3)+2*pow(x,2)+5*x+1
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Função Racional")
x = np.linspace(0.9,1.1,10000)
y= 1/(x-1)
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Função Seno")
x=np.linspace(-4,4,10000)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Função Cosseno")
x=np.linspace(-4,4,10000)
y=np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Função Seno x Função Cosseno")
x=np.linspace(-4,4,10000)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,label = "Seno")
plt.plot(x,z,label = "Cosseno")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Função Expônencial")
x = np.linspace(-10,10,10000)
y=np.exp(x)
plt.plot(x,y)
plt.show()
```

```python
'''Função sigmóide fica entre 0 e 1'''
plt.figure(figsize=(8,5))
plt.title("Função Sigmóide")
x=np.linspace(-10,10,10000)
y= 1 / (1+np.exp(-x))
plt.plot(x,y)
plt.show()
```
